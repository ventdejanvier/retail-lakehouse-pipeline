```python
from pyspark.sql.functions import col, to_timestamp, trim

# Đọc từ lớp thô (Bronze)
df_raw = spark.read.parquet("Files/bronze/raw_retail_data.parquet")

# Thực hiện làm sạch
df_silver = df_raw \
    .dropDuplicates() \
    .withColumn("Description", trim(col("Description"))) \
    .withColumnRenamed("Customer ID", "CustomerID") \
    .dropna(subset=["CustomerID", "Invoice"]) \
    .withColumn("InvoiceDate", to_timestamp(col("InvoiceDate"), "yyyy-MM-dd HH:mm:ss")) \
    .filter(col("Quantity") > 0)

# Ghi vào lớp Silver
# Ghi định dạng Parquet để bước sau đọc
df_silver.write.mode("overwrite").parquet("Files/silver/cleaned_retail_data")
```
